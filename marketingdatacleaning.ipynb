```python
import pandas as pd
df= pd.read_csv("marketing_data.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.replace(" ","")
```

```python
df.isnull().sum()
```

```python
df["Income"]=df["Income"].str.replace("$","")
```

```python
df["Income"]=df["Income"].str.replace(",","")
```

```python
df["Income"]=df["Income"].astype("float")
```

```python
df['Income']
```

```python
mean= df["Income"].mean()
```

```python
df["Income"]=df["Income"].fillna(mean)
```

```python
print(mean)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["grocerytotals"]= df[['MntWines', 'MntFruits',
       'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
       'MntGoldProds']].sum(axis=1)
```

```python
df.columns
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df["Income"], df["grocerytotals"])
plt.xlabel("Income")
plt.ylabel("grocerytotals")
```

```python
df.drop(df[df["Income"]<df["grocerytotals"]].index,inplace= True)
```

```python
df
```

```python
df.describe()[["Income","grocerytotals"]]
```

```python
q1= df['Income'].quantile(0.25)
q3= df['Income'].quantile(0.75)
iqr=q3-q1
lowerbound= q1-1.5*iqr
upperbound=q3+1.5*iqr
```

```python
df=df[(df["Income"] >= lowerbound) & (df["Income"] <= upperbound)]
```

```python
df
```

```python
q1= df['grocerytotals'].quantile(0.25)
q3= df['grocerytotals'].quantile(0.75)
iqr=q3-q1
lowerbound= q1-1.5*iqr
upperbound=q3+1.5*iqr
```

```python
df=df[(df["grocerytotals"] >= lowerbound) & (df["grocerytotals"] <= upperbound)]
```

```python
df.describe()[["Income","grocerytotals"]]
```

```python
plt.scatter(df["Income"], df["grocerytotals"])
plt.xlabel("Income")
plt.ylabel("grocerytotals")
```

```python
best_selling = df[['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']].sum()
print(best_selling)
```

```python
best_selling.plot(kind = "bar")
plt.ylabel('amount in dollars')
```

```python
df["total_purchases"]= df[['NumDealsPurchases', 'NumWebPurchases',
       'NumCatalogPurchases', 'NumStorePurchases' ]].sum(axis=1)
```

```python
avg_purchasevalue= df.grocerytotals.sum() / df.total_purchases.sum()
print("the average purchase value is ", avg_purchasevalue)
print("The average frequency of purchasing products is ", df.total_purchases.mean())
```
